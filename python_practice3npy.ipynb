Write a python program to insert a new number at the given
position
* in a list of numbers without using insert function.
* lst=[1,2,3,4,5,6,7]
* num=10
* pos=2
* [1,2,10,3,4,5,6,7]

```python
n=int(input("Enter the no. of elements in the list: "))
lst=[]
for i in range(n):
  n1=int(input("Enter elements "))
  lst.append(n1)
print("The original list: ", lst)
num=int(input("Enter the no. to be inserted in the list: "))
pos=int(input("Enter the position where the element is to be inserted: "))
#lst[pos:pos]=[num]
lst = lst[:pos] + [num] + lst[pos:]
print("The updated list is : ", lst)
```

```python
t1=tuple()
t2=tuple()
t3=tuple()
n=int(input("Enter the number of email ids you want to input: "))
for i in range(n):
  email=input("Enter the emailid: ")
  t1=t1+(email,)
  lst=email.split("@")
  t2=t2+ (lst[0],)
  t3=t3 +(lst[1],)

print("The emailids are: ", t1)
print("the usrenames are: ", t2)
print("the domain names are: ", t3)
```

write python program to create a list and search the given element in the list.
* also print the position of the number. Print error msg if the number not found
* donot use index() or find()

```python
n=int(input("Enter the number of elements in the list: "))
lst=[]
for i in range(n):
  n1=int(input("Enter the list element:"))
  lst.append(n1)
num=int(input("Enter the number to be searched:"))
flag=0
for i in range(n):
  if lst[i]==num:
    print("the number is present at ", i+1," position")
    flag=1
if flag==0:
  print("Number not found")
```
