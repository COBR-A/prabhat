```python
user_details = {
    "prabhat": "3456789",

    "user1": "123456"
}
registered_users = list(user_details.keys())
```

```python
books = [
    {"id": "00AW", "title": "Wings of Fire", "author": "A.P.J. Abdul Kalam", "price": 214.99},
    {"id": "01IM", "title": "Ignited Minds", "author": "A.P.J. Abdul Kalam", "price": 213.50},
    {"id": "02IV", "title": "India 2020: A Vision for the New Millennium", "author": "A.P.J. Abdul Kalam", "price": 212.75},
    {"id": "03MJ", "title": "My Journey: Transforming Dreams into Actions", "author": "A.P.J. Abdul Kalam", "price": 215.25},
    {"id": "04IM", "title": "The Immortals of Meluha", "author": "Amish Tripathi", "price": 314.99},
    {"id": "05SN", "title": "The Secret of the Nagas", "author": "Amish Tripathi", "price": 313.50},
    {"id": "06OV", "title": "The Oath of the Vayuputras", "author": "Amish Tripathi", "price": 312.75},
    {"id": "07SI", "title": "Scion of Ikshvaku", "author": "Amish Tripathi", "price": 315.25},
    {"id": "08SB", "title": "A Suitable Boy", "author": "Vikram Seth", "price": 215.50},
    {"id": "09MC", "title": "Midnight's Children", "author": "Salman Rushdie", "price": 514.75},
    {"id": "10WT", "title": "The White Tiger", "author": "Aravind Adiga", "price": 311.25},
    {"id": "11TG", "title": "The Guide", "author": "R.K. Narayan", "price": 510.99},
    {"id": "12TP", "title": "Train to Pakistan", "author": "Khushwant Singh", "price": 193.80},
    {"id": "13PC", "title": "Python Crash Course", "author": "Eric Matthes", "price": 225.00},
    {"id": "14IA", "title": "Introduction to Algorithms", "author": "Thomas H. Cormen", "price": 445.00},
    {"id": "15HD", "title": "Head First Design Patterns", "author": "Eric Freeman", "price": 340.00},
    {"id": "16JG", "title": "JavaScript: The Good Parts", "author": "Douglas Crockford", "price": 520.00}
]
```

```python
def login():
    while True:
        username = input("Enter username: ")
        password = input("Enter password: ")

        if username in user_details and user_details[username] == password:
            print(f"login successful! welcome, {username}.")
            return username
        else:
            print("invalid username or password. please try again.")

def register_user():
    while True:
        new_username = input("enter new username: ")

        if new_username in user_details:
            print("username already exists. please choose a different username.")
        else:
            new_password = input("enter password for the new user: ")
            user_details[new_username] = new_password
            registered_users.append(new_username)
            print(f"user '{new_username}' registered successfully!")
            return new_username

def display_books():
    print("\nbooks in library:")
    for book in books:
        print(f"id: {book['id']}, title: {book['title']}, author: {book['author']}, price: rupee {book['price']:.2f}")

def list_books_by_author_title_price():
    author_name = input("enter author's name: ")
    found_books = [book for book in books if book['author'].lower() == author_name.lower()]
    if found_books:
        print(f"\nbooks by '{author_name}':")
        for book in found_books:
            print(f"title: {book['title']}, author: {book['author']}, price: rupee {book['price']:.2f}")
    else:
        print(f"no books found for author '{author_name}'.")

def add_book():
    while True:
        try:
            title = input("enter book title: ")
            author = input("enter author's name: ")
            price = float(input("enter book price: "))
            book_id = input("enter book id: ")
        except ValueError:
            print("invalid price format. enter a valid number for the price.")
            continue

        new_book = {
            "id": book_id,
            "title": title,
            "author": author,
            "price": price
        }
        books.append(new_book)
        print(f"book '{title}' added successfully!")
        break

def count_books():
    print(f"\ntotal number of books in the library: {len(books)}")

def issue_book():
    display_books()
    book_id = input("enter the book id you want to issue: ")

    filtered_books = list(filter(lambda book:book['id'].lower == book_id.lower(), books))
    if filtered_books:
        book = filtered_books[0]
        if not book.get('issued', False):
            book['issued'] = True
            print(f"book '{book['title']}' has been issued.")
        else:
            print(f"book '{book['title']}' is already issued.")
    else:
        print("invalid book id. please enter a valid book id.")

def return_book():
    display_books()
    book_id = input("enter the book id you want to return: ")

    filtered_books = list(filter(lambda book: book['id'].lower == book_id.lower(), books))
    if filtered_books:
        book = filtered_books[0]
        if book.get('issued', False):
            book['issued'] = False
            print(f"book '{book['title']}' has been returned.")
        else:
            print(f"book '{book['title']}' is not currently issued.")
    else:
        print("invalid book id. please enter a valid book id.")

def logout(username):
    print(f"logging out {username}...")
    print("logout successful")
```

```python
def main_menu():
    current_user = None

    while True:
        print("\n********** E-Library Management System **********")
        try:
            if not current_user:
                print("1. Login")
                print("2. Register a new user")
            else:
                print(f"logged in as: {current_user}")
                print("3. display all books")
                print("4. list books by author (title and price)")
                print("5. add a book")
                print("6. count of books in library")
                print("7. issue a book")
                print("8. return a book")
                print("9. logout")

            choice = input("enter your choice: ")

            if not current_user:
                if choice == '1':
                    current_user = login()
                elif choice == '2':
                    current_user = register_user()
                else:
                    print("invalid choice. please try again.")
            else:
                if choice == '3':
                    display_books()
                elif choice == '4':
                    list_books_by_author_title_price()
                elif choice == '5':
                    add_book()
                elif choice == '6':
                    count_books()
                elif choice == '7':
                    issue_book()
                elif choice == '8':
                    return_book()
                elif choice == '9':
                    print(f"logging out {current_user}...")
                    current_user = None
                else:
                    print("invalid choice. please try again.")
        except Exception:
            print("an error occurred.")

        if current_user is None:
            break

main_menu()
```
